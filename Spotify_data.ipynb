### Making the release date as the index

```python
tracks.set_index('release_date', inplace = True)
tracks.index = pd.to_datetime(tracks.index) #this changes the release date to proper datetime object
tracks.head()
```

```python
tracks[['artists']].iloc[20]
```

```python
tracks['duration'] = tracks['duration_ms'].apply(lambda x: round(x/1000))

tracks_copy = tracks.copy()
tracks_copy.drop('duration_ms', inplace = True, axis = 1)

tracks_copy.duration.head()
```

## Correlations between the measures in the dataset

```python
corr_track= tracks_copy.drop(['key','mode','explicit'], axis = 1).corr(method='pearson')
plt.figure(figsize=(11,5))
heat_track = sns.heatmap(corr_track, annot = True, vmin = -1, vmax = 1, cmap = 'cubehelix', linewidths = 1, linecolor = 'Black')
heat_track.set_title('Correlation between measures in the data set')
heat_track.set_xticklabels(heat_track.get_xticklabels(), rotation = 90)
```

```python
sample_tracks = tracks_copy.sample(int(0.005*len(tracks_copy)))
print(len(sample_tracks))
```

```python
plt.figure(figsize = (10,5))
tracks_reg = sns.regplot(data = sample_tracks, y = 'loudness', x = 'energy', color = 'b')
tracks_reg.set_title('Loudness VS Energy Correlation')
```

```python
plt.figure(figsize = (10,5))
tracks_reg = sns.regplot(data = sample_tracks, y = 'popularity', x = 'danceability', color = 'c')
tracks_reg.set_title('Popularity VS Danceability Correlation')
```

```python
tracks_copy['dates'] = tracks_copy.index.get_level_values('release_date')
tracks_copy.dates = pd.to_datetime(tracks_copy.dates)
years = tracks_copy.dates.dt.year
```

```python
years.head()
```

## Number of songs in each year (visualized)

```python
years_hist = sns.displot(years, discrete = True, aspect = 2, height = 6, kind = 'hist').set(title = 'Number of songs released per year')
# years_hist.set_title("Number of songs released per year")
```

## Average durations of songs per year

```python
total_duration = tracks_copy.duration
fig, ax = plt.subplots(figsize = (18,5))
fig = sns.barplot(x = years, y = total_duration, ax = ax).set(title = 'year VS Duration')
plt.xticks(rotation = 90)
```

```python
total_duration_2 = tracks_copy.duration
sns.set_style(style = 'whitegrid')
fig, ax = plt.subplots(figsize = (12,6) )
fig = sns.lineplot(x = years, y = total_duration_2, ax = ax).set(title = 'Year Vs Duration ')
plt.xticks(rotation = 90)
```

## Exploring music and genre

```python
genre_tracks = pd.read_csv("C:\\Users\\princ\\OneDrive\\Documents\\datasets\\spotify\\SpotifyFeatures.csv")
```

```python
genre_tracks.head()
```

```python
plt.title('Genres of music and the duration of the songs in them')
sns.color_palette('inferno', as_cmap = True)
sns.barplot(y= 'genre', x = 'duration_ms', data= genre_tracks)
plt.xlabel('Duration in milliseconds')
plt.ylabel('Genres of music')
```

```python
plt.figure(figsize = (11,6))
sns.set_style(style = 'darkgrid')
genre = genre_tracks.sort_values('popularity', ascending = False).head(10)
genre_bar = sns.barplot(data = genre, y = 'genre', x = 'popularity')
genre_bar.set_title('Top 5 popular genres of music')
```
